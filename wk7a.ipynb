```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('sales_data.csv') 
# print(data)
```

```python
print("First 5 rows of the dataset:")
print(data.head())
```

```python
print("\nSummary statistics:")
print(data.describe())
```

```python
print("\nMissing values in each column:")
print(data.isnull().sum())
```

```python
print("\nDatatypes:")
print(data.dtypes)
```

```python
if 'Region' in data.columns:
    print("\nValue counts for Region:")
    print(data['Region'].value_counts())
```

```python
region_revenue = data.groupby('Region')['Sales_Amount'].sum()
region_revenue.plot(kind='bar', figsize=(8, 5), title='Total Sales_Amount by Region')
plt.xlabel('Region')
plt.ylabel('Sales_Amount')
plt.show()
```

```python

category_sales = data.groupby('Product_Category')['Quantity_Sold'].sum()
category_sales.plot(kind='bar', figsize=(10,6), color='purple')
plt.title('Quantity sold by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Quantity_Sold')
plt.show()
           
```

```python
plt.figure(figsize=(8,5))
plt.scatter(data['Discount'], data['Sales_Amount'], alpha=0.5, color='green')
plt.title('Discount vs Sales_Amount')
plt.xlabel('Discount')
plt.ylabel('Sales_Amount')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(data['Sales_Amount'], bins=20, color='teal', edgecolor='black')
plt.title('Distribution of Sales Amount')
plt.xlabel('Sales_Amount')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,8))
if 'Sales_Rep' in data.columns and 'Sales_Amount' in data.columns:
    sales_distribution = data.groupby('Sales_Rep')['Sales_Amount'].sum().plot(kind='pie', autopct='%1.1f%%')
    plt.title('Sales Distribution by Sales Rep')
    plt.ylabel('')
    plt.tight_layout()
    plt.show()
else:
    print("Error: 'Sales_Rep' or 'Sales_Amount' column not found in the data.")
```
